## Subtask 1: Correct years for given dataset

#### The given data is from 2022 to  2033 but the excel file has dates for year 2022 only. This seems to be a mistake . This needs to be fixed and cleaned.

```python
year = 2022

weather_df['Date'][0] = "'" + str(year) + weather_df['Date'][0][4:] + "'"
for i in range(1,len(weather_df)):
    curr_month = int(weather_df["Date"][i][5:7])
    prev_month = int(weather_df["Date"][i-1][6:8])
    if curr_month < prev_month:
        year +=1
    weather_df["Date"][i] = "'" + str(year)+weather_df['Date'][i][4:] 
    print(year)    

    
```

## Subtask 3: fix few labels in the given dataset

#### It can be clearly seen that all the labels are not fixed most of them are missing ")" in the end . we need to fix it.

```python
cols = weather_df.columns.to_list()
```

```python
cols
```

```python
for i,col in enumerate(cols):
    if '(' in col:
        cols[i] = col + ")"
        weather_df.columns = cols
        
        print(weather_df.columns)
```
